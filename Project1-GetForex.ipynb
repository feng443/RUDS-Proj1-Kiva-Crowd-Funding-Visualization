```python
url = "http://data.fixer.io/api/latest?"
```

```python
#Exchange rates delivered by the Fixer API are by default relative to EUR.

query_url = url + "access_key="+access_key

currency_response = requests.get(query_url)
currency_json = currency_response.json()
print(json.dumps(currency_json, indent=4))
```

```python
currencies = ['HNL', 'USD', 'XOF', 'PYG', 'UGX', 'PHP', 'EGP', 'KHR', 'NIO',
       'KES', 'TJS', 'IDR', 'TZS', 'BOB', 'PEN', 'GHS', 'MXN', 'PKR',
       'INR', 'HTG', 'COP', 'WST', 'GTQ', 'BIF', 'NGN', 'MZN', 'EUR',
       'RWF', 'TRY', 'SRD', 'VND', 'GEL', 'KGS', 'LRD', 'SLL', 'YER',
       'MMK', 'LAK', 'CRC', 'JOD', 'ILS', 'AMD', 'MNT', 'LSL', 'NPR',
       'AZN', 'XAF', 'ZMW', 'ALL', 'UAH', 'ZAR', 'MGA', 'MWK', 'LBP',
       'CNY', 'DOP', 'SBD', 'MDL', 'BRL', 'THB', 'SSP', 'BZD']
```

```python
rates = currency_json['rates']
```

```python
for currency in currencies:
    rate = rates.get(currency, 0)
    print(f'{currency} => {rate}')
```
